```python
import torch
w = torch.tensor(3.0, requires_grad=True)
x = torch.tensor(4.0)
y_true = torch.tensor(11.0)

lr = 0.1

y_pred = w * x
loss = (y_pred - y_true)**2
loss.backward()

print("Before Update")
print("w:", w.item())
print("loss:", loss.item())
print("Gradient:", w.grad.item())
##w.grad_zero_()
##print(w.grad)

with torch.no_grad():
  w -= lr * w.grad

print("After Update")
print("w:", w.item())
print("loss:", loss.item())
```

```python
import torch
w = torch.tensor([3.0, 1.0, 1.0, 1.0], requires_grad=True)
x = torch.tensor([4.0, 4.0, 4.0, 4.0])
y_true = torch.tensor([2.0, 4.0, 6.0, 8.0])

lr = 0.1
epochs = 10

for epoch in range(epochs):
  y_pred = w * x
  loss = ((y_pred - y_true)**2).mean()
  loss.backward()

  with torch.no_grad():
    w -= lr * w.grad
  w.grad.zero_()

  print("Epoch:", epoch + 1)
  print("w:", w)
  print("loss:", loss.item())
```

```python
import torch
import torch.nn as nn

layer = nn.Linear(3,2)
x = torch.tensor([[1.0,2.0,3.0]])

y = layer(x)
print("Layer output (y):")
print(y)

# Define an optimizer using the parameters of the 'layer'
optimizer = torch.optim.SGD(layer.parameters(), lr = 0.01)

# For demonstration, let's assume a dummy target and calculate a simple loss
y_target = torch.tensor([[0.5, -0.5]])
loss = ((y - y_target)**2).mean()

print("\nInitial loss:", loss.item())

# Perform a backward pass and optimization step
optimizer.zero_grad() # Clear previous gradients
loss.backward() # Compute gradients
optimizer.step() # Update weights

# Recalculate output and loss after one optimization step
y_after_step = layer(x)
loss_after_step = ((y_after_step - y_target)**2).mean()

print("\nOutput after one step:")
print(y_after_step)
print("Loss after one step:", loss_after_step.item())
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms

from torchvision.datasets import MNIST
from torch.utils.data import DataLoader # FIX: Corrected DataLoader import
import matplotlib.pyplot as plt

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using Device", device)

transform = transforms.ToTensor()

train_data = MNIST(
    root="data",
    train=True,
    download=True,
    transform=transform
    )

test_data = MNIST(
    root="data",
    train=False,
    download=True,
    transform=transform
)

train_loader = DataLoader(
    train_data,
    batch_size=64,
    shuffle=True
    )

test_loader = DataLoader( # FIX: Typo corrected from test_loadr
    test_data,
    batch_size=64,
    shuffle=False
    )

class DigitModel(nn.Module):
  def __init__(self):
    super().__init__()

    self.net = nn.Sequential(
        nn.Linear(28*28, 64),
        nn.ReLU(),
        nn.Linear(64, 64),
        nn.ReLU(),
        nn.Linear(64, 10)
    )

  def forward(self, x): # FIX: Corrected indentation
    return self.net(x)

# FIX: Unindented training/evaluation loop
model = DigitModel().to(device)

loss_fn = nn.CrossEntropyLoss()
optimiser = optim.Adam(model.parameters(), lr=0.001)
epochs = 10

for epoch in range(epochs):
  model.train()
  train_loss = 0.0
  for images, labels in train_loader:
    images = images.view(images.shape[0], -1).to(device)
    labels = labels.to(device)

    optimiser.zero_grad()
    output = model(images)
    loss = loss_fn(output, labels)
    loss.backward()
    optimiser.step()

    train_loss += loss.item()

  print(f"Epoch {epoch + 1}/{epochs}, Train Loss: {train_loss / len(train_loader)}")

  model.eval()
  correct = 0
  total = 0
  with torch.no_grad():
    for images, labels in test_loader: # FIX: Used test_loader
      images = images.view(images.shape[0], -1).to(device)
      labels = labels.to(device)
      outputs = model(images)
      predictions = torch.argmax(outputs, dim=1)

      total += labels.size(0)
      correct += (predictions == labels).sum().item()

    accuracy = 100 * correct / total # Calculate accuracy after iterating through test_loader
    print(f"Epoch {epoch + 1}/{epochs}, Test Accuracy: {accuracy}%")

  # These lines are currently inside the epoch loop, which means it will save and print after every epoch.
  # For a final model save and single prediction, these might be better outside the loop.
  torch.save(model.state_dict(), "digit_model.pth")
  print("Model saved successfully!")

  index = 0
  image, true_label = test_data[index]

  plt.imshow(image.squeeze(), cmap="gray") # FIX: Used imshow
  plt.title(f"True Label: {true_label}")
  plt.axis("off")
  plt.show()

  image_flat = image.view(1, -1).to(device)
  with torch.no_grad():
    model.eval()
    output = model(image_flat)
    predicted_label = torch.argmax(output, dim = 1).item() # FIX: Added output to argmax
  print("User picked Image Imdex", index)
  print(f"Predicted Label: {predicted_label}")
  print("Model Prediction", predicted_label)
  print("True Label", true_label)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms

from torchvision.datasets import MNIST
from torch.utils.data import DataLoader # FIX: Corrected DataLoader import
import matplotlib.pyplot as plt

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using Device", device)

transform = transforms.ToTensor()

train_data = MNIST(
    root="data",
    train=True,
    download=True,
    transform=transform
    )

test_data = MNIST(
    root="data",
    train=False,
    download=True,
    transform=transform
)

train_loader = DataLoader(
    train_data,
    batch_size=64,
    shuffle=True
    )

test_loader = DataLoader( # FIX: Typo corrected from test_loadr
    test_data,
    batch_size=64,
    shuffle=False
    )

class DigitModel(nn.Module):
  def __init__(self):
    super().__init__()

    self.features = nn.Sequential(
       nn.Conv2d(1, 16, 3, 1, 1),
       nn.ReLU(),
       nn.MaxPool2d(2, 2),
       nn.Conv2d(16, 32, 3, 1, 1),
       nn.ReLU(),
       nn.MaxPool2d(2, 2)
    )

    self.classifier = nn.Sequential(
    nn.Linear(32*7*7, 64),
    nn.ReLU(),
    nn.Linear(64, 10)
)

  def forward(self,x):
    x = self.features(x)
    x = x.view(x.size(0), -1)
    x = self.classifier(x)
    return x
model = DigitModel().to(device)
loss_fn = nn.CrossEntropyLoss()
optimiser = optim.Adam(model.parameters(), lr=0.001)
epochs = 10

for epoch in range(epochs):
  model.train()
  total_loss = 0

  for images, labels in train_loader:
    images = images.to(device)
    labels = labels.to(device)
    optimiser.zero_grad()
    output = model(images)
    loss = loss_fn(output, labels)
    loss.backward()
    optimiser.step()
    total_loss += loss.item()

print(f"Epoch {epoch + 1}/{epochs}, Train Loss: {total_loss / len(train_loader)}")

model.eval()
correct = 0
total = 0
with torch.no_grad():
  for images, labels in test_loader:
    images = images.to(device)
    labels = labels.to(device)
    outputs = model(images)
    predictions = torch.argmax(outputs, dim=1)

    total += labels.size(0)
    correct += (predictions == labels).sum().item()

  accuracy = 100 * correct / total
  print(f"Epoch {epoch + 1}/{epochs}, Test Accuracy: {accuracy}%")
  torch.save(model.state_dict(), "digit_model.pth")
  print("Model saved successfully!")
```
