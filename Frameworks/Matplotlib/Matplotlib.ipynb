```python
## Concept of Subplot

#plt.subplot(nrows, ncolumns,index)
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,30,40]

plt.subplot(2,2,1)
plt.plot(x,y)
plt.grid()
plt.title('Line Chart')

plt.subplot(2,2,2)
plt.bar(x,y)
plt.grid()
plt.title('Bar chart')

plt.subplot(2,2,3)
plt.scatter(x,y)
plt.grid()
plt.title('Scatter plot')

plt.subplot(2,2,4)
plt.pie(y , labels = x, autopct = '%1.1f%%', colors = ['gold', 'skyblue', 'lightgreen', 'coral'])
plt.title('Pie Chart')

plt.suptitle('Comparison of Different Plots')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
# fig, ax = plt.subplots(1,2,figsize = (10,5)) 

fig, ax = plt.subplots(1,2,figsize = (10,5)) 
x = [1,2,3,4]
y = [10,20,30,40]

ax[0].plot(x,y, color = 'orange')
ax[0].grid()
ax[0].set_title('Line Chart')

ax[1].bar(x,y, color = 'green')
ax[1].grid()
ax[1].set_title('Line Chart')

fig.suptitle('Comparison of Different Plots')

fig , ax = plt.subplots(2,1,sharex = True)
plt.tight_layout()
plt.show()
```
