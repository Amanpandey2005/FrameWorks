```python
# Merge
import pandas as pd
df_customers = pd.DataFrame({
    'CustomerID' : [1,2,3],
    'Name' : ['Ramesh', 'Suresh', 'Kalpesh']
})

df_orders = pd.DataFrame({
    'CustomerID' : [1,2,4],
    'Order_amount' : [1500, 2000, 2500]
})

df_merged = pd.merge(df_customers, df_orders, on = "CustomerID", how = "outer")
print("Outer Join")
print(df_merged)
```
