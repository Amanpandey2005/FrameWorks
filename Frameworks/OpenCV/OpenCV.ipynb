```python
## Resizing and Rescaling Images
import cv2
image = cv2.imread("OIP.jpeg")

if image is None:
    print("Error : Image is not Found")
else:
    print("Image Loaded Successfully")
    resized = cv2.resize(image,(300,300))
    cv2.imshow("Original Image",image)
    cv2.imshow("Resized image", resized)

    cv2.imwrite("resized.png",resized)
```

```python
## Cropping using Slicing 
import cv2
image = cv2.imread("OIP.jpeg")

if image is not None:
    cropped = image[100:200 , 50:150]

    cv2.imshow("Original Image",image)
    cv2.imshow("Cropped image", cropped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imwrite("cropped.png", cropped)
```

```python
import cv2
image = cv2.imread("OIP.jpeg")

if image is None:
    print("Error : Image is not Found")
else:
    print("Image Loaded Successfully")
    (h,w) = image.shape[:2]
    center = (w//2 , h//2)
    M = cv2.getRotationMatrix2D(center,90,1.0)
    rotated = cv2.warpAffine(image , M ,(h,w))
    cv2.imshow("Original Image",image)
    cv2.imshow("Rotated 90 degree image", rotated)

    cv2.imwrite("rotated.png", rotated)
```

```python
import cv2
image = cv2.imread("OIP.jpeg")
if image is None:
    print("Could not load image")
else:
    flipped_horizontal = cv2.flip(image,1)
    flipped_vertical = cv2.flip(image , 0)
    flipped_both = cv2.flip(image, -1)

    cv2.imshow("Original Image", image)
    cv2.imshow("Flipped Horizontal Image", flipped_horizontal)
    cv2.imshow("Flipped Vertical  Image", flipped_vertical)
    cv2.imshow("Both Flipped Image", flipped_both)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("OIP.jpeg")

if image is None:
    print("Error : Image is not Found")
else:
    print("Image Loaded Successfully")
    pt1 = (50,100)
    pt2 = (300,100)
    color = (255,0,0)
    thickness = 4

    cv2.line(image, pt1, pt2, color, thickness)

    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
