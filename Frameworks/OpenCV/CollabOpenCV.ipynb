```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/OIP.jpeg")

if image is None:
    print("Error : Image is not Found")
else:
    print("Image Loaded Successfully")
    pt1 = (50,100)
    pt2 = (300,100)
    color = (255,0,0)
    thickness = 4

    cv2.line(image, pt1, pt2, color, thickness)

    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/OIP.jpeg")

if image is None:
    print("Error : Image is not Found")
else:
    print("Image Loaded Successfully")
    pt1 = (50,100)
    pt2 = (300,100)
    color = (255,0,0)
    thickness = 4

    cv2.rectangle(image, pt1, pt2, color, thickness)

    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture(0)

while True :
    ret, frame = cap.read()

    if not ret:
        print("Could not Braek Frame")
        break

        cv2.imshow("Webcam feed", frame)
        if  cv2.waitKey(1) & 0xFF == ord('q'):
            print("Quitting....")
            break
            cap.release()
            cv2.destroyAllWindows()
```
